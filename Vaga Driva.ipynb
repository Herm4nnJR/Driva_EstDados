```python
import pandas as pd
```

```python
arqEmpresa = 'DadosEmpresa.csv'
arqEndereco = 'DadosEndereco.csv'
```

```python
datasetEmpresa = pd.read_csv(arqEmpresa, sep = ',', header = 0)
datasetEndereco = pd.read_csv(arqEndereco, sep = ',', header = 0)
```

```python
datasetEmpresa.head()
```

```python
datasetOpSimples = datasetEmpresa.loc[datasetEmpresa['opcao_pelo_simples'] == 'SIM']
#arquivo que contem apenas as empresas que optam pelo simples
```

```python
datasetOpSimples.head()
```

```python
datasetEndereco.head()
```

```python
Completo = pd.merge(datasetEmpresa, datasetEndereco, on = 'cnpj')
```

```python
Completo.head()
```

```python
Cwb = Completo.loc[(Completo['municipio'] == 'CURITIBA') & (Completo['capital_social'] > 5000)]
```

```python
Lond = Completo.loc[(Completo['municipio'] == 'LONDRINA') & (Completo['capital_social'] > 5000)]
```

```python
CwbLond = Cwb.append(Lond, ignore_index=True) 
#Criação do arquivo apenas com municipios de Curitiba e Londrina com capital social de mais de 5000.
```

```python
CwbLond.head()
```

```python
CwbLond.tail()
```

```python
#criando arquivos .csv
datasetOpSimples.to_csv('Simples.csv')
CwbLond.to_csv('Curitiba&Londrina_m5000.csv')
```
